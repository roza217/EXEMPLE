# Data Visualization Exercise
## CSV → Pandas → Matplotlib

Objectif : charger un fichier CSV, explorer les données et créer plusieurs graphiques.

Étapes :
1. Charger le fichier CSV
2. Explorer le DataFrame
3. Créer une nouvelle colonne
4. Créer différents graphiques


## 1. Importer les librairies
Importer :
- pandas
- matplotlib.pyplot


```python
# TODO : importer pandas et matplotlib
import pandas as pd
import matplotlib.pyplot as plt

```

## 2. Charger le fichier CSV
Le fichier s'appelle : **ventes_marketing.csv**


```python
# TODO : charger le fichier CSV dans un DataFrame appelé df
df = pd.read_csv("ventes_marketing.csv")
print(df)

```

## 3. Explorer les données

Afficher :
- les 5 premières lignes
- les statistiques descriptives


```python
# TODO : afficher les premières lignes
print(df.head())

# TODO : afficher les statistiques descriptives
print(df.describe())
```

## 4. Créer une nouvelle colonne

Créer une colonne **Total_ventes** :

Total_ventes = Produit_A + Produit_B


```python
# TODO : créer la colonne Total_ventes
df["Total_ventes"] = df["Produit_A"] + df["Produit_B"]

# afficher le DataFrame
print(df)
```

## 5. Histogramme

Créer un histogramme de **Total_ventes**.


```python
# TODO : créer un histogramme
plt.hist(df["Total_ventes"], bins=5, edgecolor="black", rwidth=0.9)
plt.title("Distribution des ventes totales")
plt.xlabel("Ventes totales")
plt.ylabel("Nombre de mois")
plt.savefig("histogramme_ventes_totales.png")
plt.show()

```

## 6. Graphique en barres

Comparer les ventes totales par mois.


```python
# TODO : créer un bar chart
plt.bar(df["Mois"], df["Total_ventes"])
plt.title("Ventes totales par mois")
plt.xlabel("Mois")
plt.ylabel("Ventes")
plt.savefig("bar_chart_ventes_totales.png")
plt.show()

```

## 7. Graphique en ligne

Comparer l'évolution des ventes :
- Produit_A
- Produit_B


```python
# TODO : créer un line chart
plt.plot(df["Mois"], df["Produit_A"], marker="o", label="Produit A")
plt.plot(df["Mois"], df["Produit_B"], marker="o", label="Produit B")
plt.title("Evolution des ventes par produit")
plt.xlabel("Mois")
plt.ylabel("Ventes")
plt.legend()
plt.savefig("line_chart_produits.png")
plt.show()

```

## 8. Nuage de points

Observer la relation entre :
- Marketing
- Total_ventes


```python
# TODO : créer un scatter plot
plt.scatter(df["Marketing"], df["Total_ventes"], s=80)
plt.title("Relation marketing / ventes")
plt.xlabel("Budget marketing")
plt.ylabel("Ventes totales")
plt.show()

```
